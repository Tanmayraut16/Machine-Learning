```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color='orange', linestyle='--', marker='8')
```

```python
# Multiple Plots
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 3, 5, 7, 11]

plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='green', marker='o')
plt.title('Plot 1')


plt.subplot(2, 2, 2)
plt.plot(y2, x, color='indigo', marker='o')
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y2, color='orange', marker='o')
plt.title('Plot 3')
```

```python
# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

# create a bar plot
plt.bar(categories, values, color='purple')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

plt.hist(data, bins=5, color='orange', edgecolor='black')
```
